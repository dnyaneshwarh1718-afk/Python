```python
import random
```

```python
jakpot = random.randint(1,100)
guess = int(input("Tell me what is the Number ? :"))
counter = 1
while guess != jakpot:
    if guess < jakpot:
        print("Try again, Guess Higher")
    else:
        print("Try again, Guess Lower")
    guess = int(input("Tell me what is the Number ? :"))
    counter += 1
print(f"You Guess it right, you take {counter} Attempt")
```
