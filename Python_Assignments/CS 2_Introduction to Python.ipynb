```python
# 1. Write a Python program to count the number 4 in a given list.  
def list_count_4(nums):
    count = 0
    for num in nums:
        if num == 4:
            count = count+1
    return count
print(list_count_4([1,4,6,7,4]))
print(list_count_4([1,4,6,4,7,4]))
```

```python
# 2. Write a Python program to test whether a passed letter is a vowel or not.  
def is_vowel(char):
    all_vowels ="aeiou"
    return char in all_vowels
print(is_vowel('c'))
print(is_vowel('e'))
```

```python
# 3. Write a Python program to check whether a specified value is contained in a group of values.  
#Test Data :
#3 -> [1, 5, 8, 3] : True
#-1 -> [1, 5, 8, 3] : False
def is_group_member(group_data, n):
    for value in group_data:
        if n == value:
            return True
    return False
print(is_group_member([1,5,8,3],3))
print(is_group_member([5,8,3],-1))
            
```

```python
# 4. Write a Python program to create a histogram from a given list of integers. 
# histogram([2, 3, 6, 5]) 
def histogram(items):
    for n in items:
        output =''
        times = n
        while(times>0):
            output +='@'
            times = times -1 
        print(output)
histogram([2, 3, 6, 5]) 
```

```python
def histogram(items):
    for n in items:
        # This tells Python: "Repeat this string n times"
        print('@' * n)
histogram([2, 3, 6, 5]) 
```

```python
# 5.Write a Python program to concatenate all elements in a list into a string and return it. 
def concat_list_data(list):
    result = ''
    for element in list:
        result += str(element)
    return result
print(concat_list_data([1,5,12,2]))
```

```python
def concat_list_data(my_list):
    # Convert all items to strings and join them with no space between
    return "".join(map(str, my_list))
print(concat_list_data([1,5,12,2]))
```

```python
# 7. Write a Python program to sum of three given integers. However, if two values are equal sum will be zero. 

a= int(input("Enter a first No: "))
b= int(input("Enter a second No: "))
c= int(input("Enter a Third No: "))

def sum_(a,b,c):
   
    if a == b or b == c or c == a:
        sum =  0
    else:
        sum = a+b+c
    return sum
print(sum_(a,b,c))
```

```python
# 8. Write a Python program to sum of two given integers. However, if the sum is between 15 to 20 it will return 20.  
a= int(input("Enter a first No: "))
b= int(input("Enter a second No: "))

def sum_(a,b):
    sum = a + b
    if 15<=sum <=20:
        return 20
    else:
        return sum
print(sum_(a,b))
```

```python
# 9. Write a Python program that will return true if the two given integer values are equal or their sum or difference is 5.  
a= int(input("Enter a first No: "))
b= int(input("Enter a second No: "))

def num_test(a,b):
    sum = a + b
    diff = a - b
    if a == b or sum == 5 or diff == 5:
        return True
    else:
        return False
print(num_test(a,b))
```
